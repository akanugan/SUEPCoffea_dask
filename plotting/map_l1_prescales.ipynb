```python
# Read the input file
with open('test.txt', 'r') as file:
    lines = file.readlines()

# Define the conditions
conditions = [
    'L1_SingleJet180/1.00',
    'L1_SingleJet200/1.00',
    'L1_HTT200er/1600.00',
    'L1_HTT240er/1.00',
    'L1_HTT270er/1.00',
    'L1_HTT280er/1.00',
    'L1_HTT300er/1.00',
    'L1_HTT320er/1.00',
    'L1_HTT380er/1.00',
    'L1_HTT360er/1.00',
    'L1_HTT400er/1.00',
    'L1_HTT450er/1.00',
    'L1_HTT500er/1.00',
    'L1_DoubleJet30_Mass_Min300_dEta_Max1p5/1.00',
    'L1_DoubleJet30_Mass_Min320_dEta_Max1p5/1.00',
    'L1_DoubleJet30_Mass_Min340_dEta_Max1p5/1.00',
    'L1_DoubleJet30_Mass_Min360_dEta_Max1p5/1.00',
    'L1_DoubleJet30_Mass_Min380_dEta_Max1p5/1.00',
    'L1_DoubleJet30_Mass_Min400_dEta_Max1p5/1.00'
]

# Create a dictionary to store runs for each condition
runs_dict = {condition: [] for condition in conditions}

# Process the lines to extract the runs for each condition
for line in lines:
    parts = line.strip().split(',') 
    run = parts[0]
    conditions_in_line = parts
    #print(parts)
    for condition in conditions:
        if condition in conditions_in_line:
            runs_dict[condition].append(run)

# Output the run numbers list for each condition
for condition, runs_list in runs_dict.items():
    print(f"Runs for condition {condition}: {', '.join(runs_list)}")

    
```
