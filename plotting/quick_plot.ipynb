```python
_, SR_exp = ABCD_9regions_errorProp(h, xregions, yregions, sum_var='x')

A, B, C, D, E, F, G, H, SR, _ = ABCD_9regions(
        h, xregions, yregions, sum_var='x', return_all=True
    )

for i, (region_name, region) in enumerate(zip(['A','B','C','D','E','F', 'G','H','SR'], [A, B, C, D, E, F, G, H, SR])):

    # Define the variables
    a, b, c, d, e, f, g, h, sr = sp.symbols('A B C D E F G H SR')

    vars = [a, b, c, d, e, f, g, h, sr]

    # Define the expression
    alpha = vars[i] / (a+b+c+d+e+f+g+h+sr)

    # Define the errors associated with each variable
    delta_A, delta_B, delta_C, delta_D, delta_E, delta_F, delta_G, delta_H, delta_SR = sp.symbols('delta_A delta_B delta_C delta_D delta_E delta_F delta_G delta_H delta_SR')

    # Define the values and errors for each variable
    values = [A.sum().value,
        B.sum().value,
        C.sum().value,
        D.sum().value,
        E.sum().value,
        F.sum().value,
        G.sum().value,
        H.sum().value,
        SR.sum().value]
    errors = [A.sum().variance,
        B.sum().variance,
        C.sum().variance,
        D.sum().variance,
        E.sum().variance,
        F.sum().variance,
        G.sum().variance,
        H.sum().variance,
        SR.sum().variance]

    # Substitute the values and errors into the expression
    substituted_expr = alpha.subs([(a, values[0]), (b, values[1]), (c, values[2]), (d, values[3]), (e, values[4]), (f, values[5]), (g, values[6]), (h, values[7]), (sr, values[8]),
                                (delta_A, errors[0]), (delta_B, errors[1]), (delta_C, errors[2]), (delta_D, errors[3]), (delta_E, errors[4]), (delta_F, errors[5]), (delta_G, errors[6]), (delta_H, errors[7]), (delta_SR, errors[8])])

    # Calculate the propagated error (delta_alpha)
    partial_derivatives = [sp.diff(alpha, var) for var in (a, b, c, d, e, f, g, h, sr)]
    partial_derivative_values = [deriv.subs([(a, values[0]), (b, values[1]), (c, values[2]), (d, values[3]), (e, values[4]), (f, values[5]), (g, values[6]), (h, values[7]), (sr, values[8])]) for deriv in partial_derivatives]
    delta_alpha = sp.sqrt(sum([(partial_derivative_values[i])**2 * errors[i] for i in range(len(partial_derivatives))]))

    # Evaluate the expression and propagated error
    result = sp.N(substituted_expr)
    propagated_error = sp.N(delta_alpha)

    print(region_name, round(result, 5), '+-', round(propagated_error, 5))

if rebin:
    SR = rebin_piecewise(SR, new_bins)
    SR_exp = rebin_piecewise(SR_exp, new_bins)

print(SR.values()/SR_exp.values())

fig, axs = plot_ratio([SR_exp, SR], labels=["$SR_{exp}$", "$SR_{obs}$"])
axs[0].set_title(r"ISR Closure - 2018, $\eta$ > 0")
# _ = axs[0].plot([],[],' ',label='KS p-value: {}'.format(round(ks.pvalue,3)))
_ = axs[1].set_ylim(0.5, 2)
# _ = axs[0].legend()
#print(axs[1].lines[0].get_ydata())
#print(axs[0].lines[0].get_ydata())
```
